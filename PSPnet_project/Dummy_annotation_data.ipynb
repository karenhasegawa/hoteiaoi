```python
import os 
import os.path as osp

import cv2
import numpy as np

import torch
```

```python
rootpath = '/content/drive/MyDrive/PSPNet/data_dataset_voc_20220610_(1)'
rootpath_1 = '/content/drive/MyDrive/PSPNet/data_dataset_voc_20220610_(1)/SegmentationClassPNG'
rootpath_2 = '04_10m'
```

```python
# ダミーデータの作成
batch_size = 2
dummy_img = torch.rand(batch_size, 3, 475, 475)
print(dummy_img)
dummy_img.save('/content/drive/MyDrive/PSPNet/testimg')
```

```python
if osp.join(rootpath_1,rootpath_2, '%s.PNG'):
    imgpath_template = osp.join(rootpath_1,rootpath_2, '%s.PNG')

val_id_names = osp.join(rootpath + '/ImageSets/Segmentation/val.txt')
val_img_list = list()

for line in open(val_id_names):
  file_id = line.strip()  # 空白スペースと改行を除去
  img_path = (imgpath_template % file_id)  # 画像のパス
  val_img_list.append(img_path)
```

```python
#ブランク画像
height = 3000
width = 4000
blank = np.zeros((height, width, 3))

os.mkdir(rootpath_1+'/'+rootpath_2)

for i in val_img_list:
  cv2.imwrite(i,blank)
```
